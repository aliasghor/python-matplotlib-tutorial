# Pengenalan Matplotlib
- Matplotlib adalah salah satu module dalam bahasa pemograman Python yang dapat memvisualisasikan data dalam bentuk graphs.
- Dengan Matplotlib, graphs akan digambarakan dalam suatu `figure` yang di dalamnya dapat terdiri dari satu atau lebih `axes`.

Logo reference: [numerique.ostralo.net](https://numerique.ostralo.net/python_matplotlib/images/logo.png).

# Import modules

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Membuat plotting sederhana
Membuat sebuah `figure` yang memiliki sebuah `axes`; melakukan plotting data pada `axes`.

```python
# Persiapan sample data
x = [1,2,3,4]
y = [1,4,2,3]

fig, ax = plt.subplots() # Membuat sebuah object figure dan sebuah axes

ax.plot(x,y) # Melakukan plotting data pada object axes
```

Alternatif lain, kita bisa langsung memanfaatkan fungsi `plot()` pada `pyplot` untuk melakukan plotting sederhana.

```python
# Persiapan sample data
x = [1,2,3,4]
y = [1,4,2,3]

plt.plot(x,y)
```

# Komponen dari sebuah Matplotlib `figure`
Perhatikan perbedaan antara `axes` dan `axis`!


Apa itu `figure`, `axes`, dan `axis`?
- **Figure**: `figure` adalah sebuah object untuk menampung sebuah object `axes` (gambar), dalam sebuah figure kita bisa dapat mempunyai 1 atau lebih `axes`.

- **Axes**: `axes`adalah sebuah object area gambar sewkatu kita memvisualisasi data, `axes` akan selalu berada didalam sebuah object `figure`.

- **Axis**: `axis` mengidentifikasi suatu sumbu dalam object `axes`. Jumlah `axis` sendiri bisa terdiri dari 2 atau 3. Jika kita memvisualisasikan data 2 dimensi maka akan terdiri dari 2 axis, yaitu: sumbu X, dan sumbu Y. Jika kita memvisualisasikan data 3 dimensi maka akan terdiri dari 3 axis, yaitu: sumbu X, sumbu Y, dan sumbu Z.

# Dua cara dalam menggunakan Matplotlib
Pada dasarnya terdapat dua cara untuk menggunakan Matplotlib, yaitu:
- OO Style (Object-Oriented Style)
- `pyplot` style

# Mengenal OO Style (Object-Oriented Style) untuk melakukan plotting data
Secara eksplisit membuat object `figure` beserta `axes`, dan memanggil methods dari keduanya.

```python
# Persiapan data
x = np.linspace(0,2,100)
x
```

```python
fig, ax = plt.subplots() # Membuat sebuah object figure dan axes

# Melakukan tiga plotting data pada object axes
ax.plot(x,x,label="linear")
ax.plot(x,x**2,label="quadratic")
ax.plot(x,x**3,label="cubic")

ax.set_xlabel("Sumbu X") # Menyertakan x-label pada axes
ax.set_ylabel("Sumbu Y") # Menyertakan y-label pada axes
ax.set_title("Plot sederhana menggunakan OO Style (Object-Oriented Style)") # Menyeratakan title pada axes
ax.legend() # Menyertakan legend pada axes
```

# Mengenal pyplot Stlye untuk melakukan plotting data
Mengandalkan `pyplot` untuk membuat dan mengelola `figures` dan `axes`, serta menggunakan fungsi dari `pyplot` untuk melakukan plotting data.

```python
x = np.linspace(0,2,100)
x
```

```python
# Plotting tiga variant data pada axes
plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quadratic")
plt.plot(x,x**3,label="cubic")

plt.xlabel("Sumbu X") # Menyeratakkan x-label pada axes
plt.xlabel("Sumbu Y") # Menyeratakkan y-label pada axes
plt.title("Plot sederhana menggunakan pyplot style") # Menyertakan title pada axes

plt.legend() # Menyertakan legend pada axes
```
